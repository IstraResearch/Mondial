```python
def solve(num_balls: int, weigh_func):
    """
    קוד לדוגמה
    הקוד אינו יעיל ואינו תומך ב- 3 כדורים כנדרש אבל אפשר לשחק איתו
    """

    def check_quartet(first, weigh):
        res1, res2 = weigh(first, 2, first + 2, 2)
        if res1 == res2:
            return 0
        res3, dummy = weigh(first, 4, first + 3, 0)
        if res2 > 0:
            first += 2;
            res1 = 3 - res2
        if res1 != res3:
            first += 1;
        return first

    assert num_balls > 2, f"Error: Number of balls is too small ({num_balls})"
    assert balls != 3, "Sorry, The case of 3 balls is not implemented yet"
    # iterate over quartets (first ball in each quartet)
    for i in range(1, num_balls + 1, 4):
        # fix the last quartet to start at num_balls-3
        first = min(i, num_balls - 3)  
        res = check_quartet(first, weigh_func)
        if res > 0:
            break
    assert res > 0, "Error: all balls are equal"
    return res
```

<h4 style="text-align: right; direction: rtl; float: right; clear: both;">
פונקציית שקילה לדוגמה:
</h4>

```python
def get_weigh_func(balls, fake_id, is_heavy):
    def _weigh(s, n):
        assert n >= 0 and n % 2 == 0 and s > 0 and s + n <= balls + 1
        if n == 0:
            return 0
        if fake_id < s:
            return 0
        elif fake_id < s + n//2:
            return 1 if is_heavy else 2
        elif fake_id < s + n:
            return 2 if is_heavy else 1
        return 0

    def weigh(s1: int, n1: int, s2: int, n2: int):
        if n1 > 0 and n2 > 0:
            assert s1 + n1 <= s2 or s2 + n2 <= s1
        return _weigh(s1, n1), _weigh(s2, n2)

    return weigh
```
